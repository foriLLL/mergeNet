## 绘制准确率

```python
file = './log/lr0.03.log'
res = []
with open(file) as f:
    line = f.readline()
    while line:
        splitted = line.split('accuracy : ')
        if len(splitted) > 1:
            num = float(splitted[1][:-2])
            res.append(num)
        line = f.readline()

import matplotlib.pyplot as plt
plt.xlabel('epoch')
plt.ylabel('precision')
plt.plot(res)
```

## 绘制 Loss

```python
res = []
with open(file) as f:
    line = f.readline()
    while line:
        splitted = line.split('average loss : ')
        if len(splitted) > 1:
            num = float(splitted[1].split(' , ')[0])
            res.append(num)
        line = f.readline()

import matplotlib.pyplot as plt
plt.xlabel('epoch')
plt.ylabel('loss')
plt.plot(res)
```
